```python
import csv
import numpy as np
from prettytable import PrettyTable
```

```python
def get_data(file,option=1,QorM=1):
    '''Accepts the file name of a weather data file and returns a list of lists containing the dataset'''
    with open(file,newline='') as f:
        reader = csv.reader(f)
        weather_data = list(reader)
    for a in range(1,len(weather_data)):
        for b in [2,3]:
            weather_data[a][b] = int(weather_data[a][b])
        for c in [4,5]:
            if weather_data[a][c]=='0':
                weather_data[a][c] = int(weather_data[a][c])
            else:
                weather_data[a][c] = float(weather_data[a][c])    
    return weather_data
```

```python
def print_data(file):
    '''Accepts a list of lists and returns a Pretty Table of the data'''
    p.clear_rows()
    p.field_names = [i for i in file[0]]
    for i in range(1,len(file)):
        p.add_row(file[i])
    for j in range(len(file[0])):
        if j in [2,3,4,5]:
            p.align[file[0][j]] = 'r'
    print(p)
```

```python
def get_all_data(file):
    '''Accepts a list of lists and returns a list with summary calculations appended to the end of the list'''
    x = get_data(file)
    x.append(calc_averages(x))
    return x
```

```python
def get_quarter_data(file):
    '''Accepts a list of lists and returns a list with summary calculations (by quarter) appended to the end of the list. Also displays a menu to choose the quarter'''
    print('1. First Quarter: January - March\n2. Second Quarter: April - June\n3. Third Quarter: July - September\n4. Fourth Quarter: October - December')
    quarter = int(input('Choose the quarter: '))
    
    if quarter==1:
        x = get_data(file)[0:91]
    elif quarter==2:
        x = [get_data(file)[0]] + get_data(file)[91:182]
    elif quarter==3:
        x = [get_data(file)[0]] + get_data(file)[182:274]
    else:
        x = [get_data(file)[0]] + get_data(file)[274:366]
    
    x.append(calc_averages(x))
    return x
```

```python
def get_month_data(file):
    '''Accepts a list of lists and returns a list with summary calculations (by month) appended to the end of the list. Also presents a menu to choose the month'''
    p.clear_rows()
    print('1. January           5. May              9. September\n2. February          6. June             10. October\n3. March             7. July             11. November\n4. April             8. August           12. December')
    month = int(input('Choose the month: '))
    
    if month==1:
        x = get_data(file)[0:32]
    elif month==2:
        x = [get_data(file)[0]] + get_data(file)[32:60]
    elif month==3:
        x = [get_data(file)[0]] + get_data(file)[60:91]
    elif month==4:
        x = [get_data(file)[0]] + get_data(file)[91:121]
    elif month==5:
        x = [get_data(file)[0]] + get_data(file)[121:152]
    elif month==6:
        x = [get_data(file)[0]] + get_data(file)[152:182]
    elif month==7:
        x = [get_data(file)[0]] + get_data(file)[182:213]
    elif month==8:
        x = [get_data(file)[0]] + get_data(file)[213:244]
    elif month==9:
        x = [get_data(file)[0]] + get_data(file)[244:274]
    elif month==10:
        x = [get_data(file)[0]] + get_data(file)[274:305]
    elif month==11:
        x = [get_data(file)[0]] + get_data(file)[305:335]
    elif month==12:
        x = [get_data(file)[0]] + get_data(file)[335:366]
        
    x.append(calc_averages(x))
    return x
```

```python
def calc_averages(file):
    '''Accepts a list of lists and returns the summary statistics for specific columns'''
    tmax_avg = f'{np.mean([row[2] for row in file][1:]):.2f}'
    tmin_avg = f'{np.mean([row[3] for row in file][1:]):.2f}'
    prcp_avg = f'{np.mean([row[5] for row in file][1:]):.2f}'
    return ['','',tmax_avg,tmin_avg,'',prcp_avg]
```

```python
def display_menu(menu=1):
    '''Displays the main menu for the Weather Reporter Application'''
    print('      Weather Reporter        \n============================\n1. Print Table (Entire Year)\n2. Print Table (by Quarter)\n3. Print Table (by Month)')
    print()
```

```python
def main():
    '''Executes the application loop. Prompts the user for the name of the file to view, then displays menu for various viewing options. Continues to prompt for menu choice until user enters -1. To run this program, you would have a single function call to main()'''
    print('Welcome to the Weather Reporter!')
    file = input('Enter name of the weather data file: ')
    display_menu()
    option = int(input('Choose option (-1 to quit): '))
    
    while option!=-1:
        if option==1:
            print_data(get_all_data(file))
        elif option==2:
            print_data(get_quarter_data(file))
        elif option==3:
            print_data(get_month_data(file))
        display_menu()
        option = int(input('Choose option (-1 to quit): '))
    
    if option==-1:
        print('Thanks for using Weather Reporter')
```
